```python
import requests
from bs4 import BeautifulSoup
URL="https://www.flipkart.com/apple-iphone-15-blue-128-gb/p/itmbf14ef54f645d?pid=MOBGTAGPAQNVFZZY&lid=LSTMOBGTAGPAQNVFZZYSCIIOB&marketplace=FLIPKART&q=iphone+15&store=tyy%2F4io&srno=s_1_1&otracker=search&otracker1=search&fm=organic&iid=e0b6f198-d7c6-4082-8ece-f51a0bba8e97.MOBGTAGPAQNVFZZY.SEARCH&ppt=hp&ppn=homepage&ssid=z7a4xmwwrk0000001700557515774&qH=2f54b45b321e3ae5"
```

```python
r=requests.get(URL)
htmlcontent=r.content
print(htmlcontent)
```

```python
soup=BeautifulSoup(htmlcontent,"html.parser")
print(soup)
print(soup.prettify)
```
